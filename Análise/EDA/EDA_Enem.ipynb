```python
# Using Dask to read the data
enem = dd.read_csv(
    '../../Dados/Enem/microdados_enem_2019/DADOS/MICRODADOS_ENEM_2019.csv',
    blocksize = 16 * 1024 * 1024,
    usecols = cols_dict.keys(),
    sep = ';',
    header = 0,
    encoding='latin-1',
    dtype={'NO_MUNICIPIO_ESC': 'object',
       'SG_UF_ESC': 'object'}
)
```
